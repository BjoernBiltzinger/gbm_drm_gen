```python
%matplotlib inline
%matplotlib notebook
from gbm_drm_gen import *
from threeML.plugins.OGIP.response import OGIPResponse

import astropy.io.fits as fits
```

## Setup

If all the variables are setup, then one simply needs to specify the files and go:

```python
n6 = DRMGenTTE('glg_tte_n6_bn110721200_v00.fit',
               trigdat='glg_trigdat_all_bn110721200_v01.fit',
               mat_type=2,
               cspecfile='glg_cspec_n6_bn110721200_v00.pha')
```

## DRM to FITS file
To make an OGIP compliant DRM, simply specify the Ra and Dec and a file name:


```python
n6.to_fits(ra=0,dec=0,'test',overwrite=True)
```

You can read your DRM with 3ML or XSPEC:

```python
ogip_rsp = OGIPResponse('test_n6.rsp')
```
