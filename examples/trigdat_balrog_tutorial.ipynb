```python
def sample_2dgauss(phi_avg, theta_avg):
    #phi_avg = 0.0
    #theta_avg = 0.0
    sigma_phi = 2.0
    sigma_theta = 1.0
    
    A = 1/(2*np.pi*sigma_phi*sigma_theta)
    phi = np.random.normal(phi_avg,sigma_phi)
    theta = np.random.normal(theta_avg,sigma_theta)
    F = A*np.exp(-((phi-phi_avg)**2)/(2*sigma_phi) )*np.exp(-((theta-theta_avg)**2)/(2*sigma_theta) )
    return phi,theta,F
```

```python
delta_list = []
real_coord = SkyCoord(0.,0.,frame='icrs',unit='degree')
    

for i in range(0,100):
    phi_list = []
    theta_list = []    
    phi_avg = np.random.normal(0,0.5)
    theta_avg = np.random.normal(0,0.25)
    
    for k in range(0,200):
        phi, theta, P = sample_2dgauss(phi_avg,theta_avg)
        phi_list.append(phi)
        theta_list.append(theta)
        
    coord_list = SkyCoord(phi_list,theta_list,frame='icrs',unit='degree')
    delta_list.extend(coord_list.separation(real_coord).value)
```

```python
def double_gauss_pdf(phi,theta):
    phi_avg = 0.0
    theta_avg = 0.0
    sigma_p = 2.0
    sigma_t = 1.0
    A = 1/(2*np.pi*sigma_p*sigma_t)
    F = A*np.exp(-((phi-phi_avg)**2)/(2*sigma_p) )* np.exp(-((theta-theta_avg)**2)/(2*sigma_t) )
    return F
```

```python
deltas=[]
deltas_z=[]
z_tot = []

for g in range (0,20):

    x = []
    y = []
    z = []

    for i in range (0,61):
        for k in range (0,31):
            phi = -6.0 + 6.0*i/30
            theta = -3.0 + 3.0*k/15

            x.append(phi)
            y.append(theta)
            z.append(double_gauss_pdf(phi,theta))


    real_coord = SkyCoord(np.random.normal(0,1),np.random.normal(0,3),frame='icrs',unit='degree')
    coord_list = SkyCoord(x,y,frame='icrs',unit='degree')
    temp = coord_list.separation(real_coord).value
    deltas.extend(temp)
    deltas_z.extend(np.multiply(temp,z))
    z_tot.extend(z)
```

```python
bin_d = np.linspace(0,20.0,201)
```

```python
z_binned=np.zeros((len(bin_d)))

for i,delta in enumerate(deltas):
    z_binned[np.digitize(delta, bin_d)] += z_tot[i] 
    
```

```python
ax, fig = plt.subplots(figsize=(10,6))
plt.scatter(bin_d, z_binned)
```
